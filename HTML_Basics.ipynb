<b>Question 1 : </b>Explain the basic structure of an HTML page. Why is it important to follow this structure? <br>

<b>Answer:</b> The basic structure for an html page is:-


```python
<!DOCTYPE html>
<html lang=" ">
<head>
<title> Hello, World!</title>
</head>
<body>
<h1>Hello, World!</h1>
<p>This is my first HTML page.</p>
</body>
</html>
```

Why is it important to follow this structure:- <br> <br>
- <b>Browser Rendering</b>: Modern web browsers rely on this consistent structure to correctly parse and display the content as intended by the developer.<br> <br>

- <b>Search Engine Optimization (SEO)</b>: Search engines use the HTML structure and semantic tags to understand the page's content hierarchy and relevance, which helps in proper indexing and ranking.<br> <br>

- <b>Accessibility: </b>Assistive technologies, such as screen readers used by people with visual impairments, depend on a logical HTML structure to navigate and vocalize content in a meaningful order.<br> <br>

- <b>Maintainability and Collaboration:</b> A consistent and logical structure makes the code easier for developers to read, understand, troubleshoot, and maintain, especially when working in teams.<br> <br>

- <b>Separation of Concerns:</b> It helps maintain the separation of concerns, where HTML manages the structure, CSS manages the presentation, and JavaScript manages the behavior.<br>

<b>Question 2:</b> You’re building a small personal profile page. Which HTML tags will you use to highlight your name as a heading, describe yourself in a paragraph and emphasize a key skill? <br>

<b>Answer:</b> To build a small personal profile page, you would use the following HTML tags: <br> <br>

- <b>To highlight your name as a heading:</b>
The < h1 > tag should be used for your name, as it represents the main heading for the page content, which in this case is your profile. <br> <br>

- <b>To describe yourself in a paragraph:</b>
The < p > tag is the standard way to define a paragraph of text, perfect for a brief self-description.<br> <br>

- <b>To emphasize a key skill:</b>
The < strong > tag is appropriate for emphasizing a key skill, indicating that the text has strong importance or seriousness. The browser will typically render this in bold text by default. <br> <br>

Here is an example of how these tags would be used:


```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Profile Page</title>
</head>
<body>
    <section>
        <h1>DIVYANSHI SEN</h1>
        <P>I am a student of PW SKILLS. <br> I am learning Full Stack Devlopment From this platform.</P>
        <p>Key Skill: <strong>Good Communication</strong></p>
    </section>
</body>
</html>
```

<b>Question 4:</b> You’re creating a form that asks the user to enter their name, select their country from a list, and click a button to submit. Write the HTML code for this. <br>
<b>Answer :-</b>

```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Profile Page</title>
</head>
<body>
  <h1> Submit Your Details </h1>
  <form>
    <div>
        <label for="Full_Name">Full Name</label><br>
        <input type="text" id="Full_Name" name="Full_Name" placeholder="Enter your Full Name" required/>
        </div>
        <br>
        <div>
            <label for="Country">Select Your Country</label><br>
            <select name="Country" id="Country">
                <option value="India">India</option>
                <option value="USA">United States</option>
                <option value="Canada"> Canada</option>
                <option value="China">China</option>
                <option value="Bangladesh">Bangladesh</option>
                <option value="Other">Other</option>
            </select>
    </div>
    <br>
        <div>
            <button type="submit">Submit</button>
        </div>
  </form>
</body>
</html>
```
